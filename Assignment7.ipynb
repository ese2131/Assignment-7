### 1) Use Pandas’ read_csv function directly on this URL to open it as a DataFrame

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
url = 'http://www.ldeo.columbia.edu/~danielmw/usgs_earthquakes_2014.csv'
df = pd.read_csv(url)
df
```

### 2) Re-read the data in such a way that all date columns are identified as dates and the earthquake ID is used as the index

```python
df = pd.read_csv(url,
                 parse_dates=['time', 'updated'],
                 index_col='id')
df
```

### 3) Use describe to get the basic statistics of all the columns

```python
df.describe()
```

### 4) Use nlargest to get the top 20 earthquakes by magnitude

```python
top20 = df.nlargest(20, 'mag')
top20[['time', 'mag', 'place', 'depth']]
```

### 5) Extract the state or country using Pandas text data functions

```python
df['country'] = df['place'].str.split(', ').str.get(-1)
```

### 6) Display each unique value from the new column

```python
df['country'].unique()
```

### 7) Create a filtered dataset that only has earthquakes of magnitude 4 or larger

```python
df_filtered = df[df['mag'] >= 4]
df_filtered
```

### 8) Using the filtered dataset (magnitude > 4), count the number of earthquakes in each country/state. Make a bar chart of this number for the top 5 locations with the most earthquakes

```python
# Creating country counts
country_counts = df_filtered['country'].value_counts()
top5 = country_counts.nlargest(5)

# Plot
fig, ax = plt.subplots(figsize=(9, 5))
top5.plot(kind='bar', ax=ax, color='steelblue', edgecolor='white')
ax.set_title('Top 5 Locations by Earthquake Count', fontsize=14)
ax.set_xlabel('Location (Country / US State)', fontsize=12)
ax.set_ylabel('Number of Earthquakes', fontsize=12)
ax.tick_params(axis='x', rotation=30)

plt.tight_layout()
plt.show()
```

### 9) Make a histogram the distribution of the Earthquake magnitudes

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5), sharey=False)

bins = np.arange(-2, 9, 0.5)

# Unfiltered
axes[0].hist(df['mag'].dropna(), bins=bins, color='steelblue', edgecolor='white')
axes[0].set_yscale('log')
axes[0].set_title('All Earthquakes (log scale)', fontsize=13)
axes[0].set_xlabel('Magnitude', fontsize=11)
axes[0].set_ylabel('Count (log scale)', fontsize=11)

# Filtered (mag >= 4)
axes[1].hist(df_filtered['mag'].dropna(), bins=bins, color='tomato', edgecolor='white')
axes[1].set_yscale('log')
axes[1].set_title('Magnitude ≥ 4 Only (log scale)', fontsize=13)
axes[1].set_xlabel('Magnitude', fontsize=11)
axes[1].set_ylabel('Count (log scale)', fontsize=11)

plt.suptitle('Distribution of Earthquake Magnitudes', fontsize=15, y=1.02)
plt.tight_layout()
plt.show()
```

### 10) Visualize the locations of earthquakes by making a scatterplot of their latitude and longitude

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Unfiltered
sc1 = axes[0].scatter(
    df['longitude'], df['latitude'],
    c=df['mag'], cmap='plasma',
    s=3, alpha=0.4, linewidths=0
)
plt.colorbar(sc1, ax=axes[0], label='Magnitude')
axes[0].set_title('All Earthquakes', fontsize=13)
axes[0].set_xlabel('Longitude', fontsize=11)
axes[0].set_ylabel('Latitude', fontsize=11)
axes[0].set_facecolor('#1a1a2e')

# Mag >4
sc2 = axes[1].scatter(
    df_filtered['longitude'], df_filtered['latitude'],
    c=df_filtered['mag'], cmap='plasma',
    s=10, alpha=0.7, linewidths=0
)
plt.colorbar(sc2, ax=axes[1], label='Magnitude')
axes[1].set_title('Magnitude ≥ 4 Only', fontsize=13)
axes[1].set_xlabel('Longitude', fontsize=11)
axes[1].set_ylabel('Latitude', fontsize=11)
axes[1].set_facecolor('#1a1a2e')

plt.suptitle('Global Earthquake Locations (2014)', fontsize=15)
plt.tight_layout()
plt.show()
```
